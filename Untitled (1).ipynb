```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm
import numpy as np
import patsy
from sklearn import linear_model
from sklearn import metrics
```

```python
Adresses = pd.read_csv("Addresses.csv")
```

```python
panama = Adresses[Adresses["sourceID"] == "Panama Papers"]

panama;
```

```python
num_adr = panama["countries"].value_counts()

fig, ax = plt.subplots()

num_adr.head(20).plot(kind = "bar", ax=ax)

fig.tight_layout()
```

```python
numadr = pd.DataFrame(num_adr)

numadr = numadr.rename(columns = {"countries":"# of Entities"})

CCodes = pd.read_csv("https://gist.githubusercontent.com/tadast/8827699/raw/f5cac3d42d16b78348610fc4ec301e9234f82821/countries_codes_and_coordinates.csv")

CCodes["Code"] = CCodes["Alpha-3 code"].str.replace('"', '').str.replace(' ', '')

c_list = ["Alpha-2 code", "Alpha-3 code", "Numeric code", "Latitude (average)", "Longitude (average)"]

CCs = CCodes.drop(c_list, axis = 1)

numadr = pd.merge(numadr, CCs, left_index = True, right_on = "Country")

numadr
```

```python
pop_raw = pd.read_csv("world_population.csv")

pop_raw

numadr_pop = pd.merge(numadr, pop_raw, left_on = "Code", right_on = "CCA3")

df = numadr_pop.drop(columns = ["Rank", "CCA3", "Country/Territory", "Capital",
                           "2022 Population", "2020 Population", "2010 Population",
                           "2000 Population", "1990 Population", "1980 Population",
                           "1970 Population", "Area (km²)", "Density (per km²)",
                           "Growth Rate", "World Population Percentage"]
               )

df
```

```python
CPI = pd.read_csv("CPI_history.csv")

CPI["CPI_avg"] = CPI[["CPI 2015 Score", "CPI 2014 Score", "CPI 2013 Score", "CPI 2012 Score"]].mean(axis = 1)
CPI
```

```python
merge = pd.merge(df, CPI, left_on = "Code", right_on = "Country Code")

df = merge.drop(columns = ["Continent", "CPI 2016 Rank", "Country_y", "Country Code",
                           "CPI 2016 Score", "CPI 2015 Score", "CPI 2014 Score", "CPI 2013 Score",
                           "CPI 2012 Score"]
               )

df
```

```python
ineq_raw = pd.read_csv("combined_final_last_10_years.csv") # no codes

ineq_2015 = ineq_raw[ineq_raw["year"] == 2015]

name_changes = {"Cote d'Ivoire":"Côte d'Ivoire", "Kyrgyz Republic":"Kyrgyzstan", "Lao":"Lao People's Democratic Republic",
                "Palestine":"Palestinian Territory, Occupied", "Moldova":"Moldova, Republic of", 
                "North Macedonia":"Macedonia, the former Yugoslav Republic of", "Slovak Republic":"Slovakia"
               }

ineq_2015 = ineq_2015.replace(name_changes)

ineq_merge = pd.merge(ineq_2015, CCs, left_on = "country", right_on = "Country", how = "left")

ineq_merge
```

```python
ineq_avg_raw = ineq_raw.pivot_table(index = "country", 
                     values = ["demox_eiu", "income_per_person",
                                "invest_%_gdp", "tax_%_gdp",
                                "gini_index"],
                               )

name_changes = {"Cote d'Ivoire":"Côte d'Ivoire", "Kyrgyz Republic":"Kyrgyzstan", "Lao":"Lao People's Democratic Republic",
                "Palestine":"Palestinian Territory, Occupied", "Moldova":"Moldova, Republic of", 
                "North Macedonia":"Macedonia, the former Yugoslav Republic of", "Slovak Republic":"Slovakia",
                "Congo, Dem. Rep.":"Congo, the Democratic Republic of the","Congo, Rep.":"Congo",
                "Iran":"Iran, Islamic Republic of","Syria":"Syrian Arab Republic",
                "Tanzania":"Tanzania, United Republic of"
               }


ineq_avg_raw = ineq_avg_raw.reset_index()

ineq_avg = ineq_avg_raw.replace(name_changes)

ineq_merge = pd.merge(ineq_avg, CCs, left_on = "country", right_on = "Country", how = "left")

ineq_merge = ineq_merge.drop(columns = ["Country"])

df_merge = pd.merge(df, ineq_merge, on = "Code", how = "left")

df_merge = df_merge.drop(columns = ["country"])

df_merge["Entities per Capita"] = df_merge["# of Entities"]/df_merge["2015 Population"]

df_merge["ln_entities_per_cap"] = np.log(df_merge["Entities per Capita"])

df_merge["ln_income"] = np.log(df_merge["income_per_person"])

df_merge["ln_pop"] = np.log(df_merge["2015 Population"])

fig, ax = plt.subplots(3)

df_merge.plot(x = "CPI_avg", y = "ln_entities_per_cap", kind = "scatter", ax=ax[0])

df_merge.plot(x = "income_per_person", y = "ln_entities_per_cap", kind = "scatter", ax=ax[1])

df_merge.plot(x = "ln_income", y = "ln_entities_per_cap", kind = "scatter", ax=ax[2])

plt.tight_layout()
```

```python
df_merge
```

```python
df = df_merge.drop(columns = ["2015 Population", "# of Entities", "Entities per Capita"])

df = df.rename(columns = {"invest_%_gdp":"invest_pct_gdp", "tax_%_gdp":"tax_pct_gdp"})
df.info()
```

```python
y, X = patsy.dmatrices("ln_entities_per_cap ~ CPI_avg + C(Region)", df, return_type='matrix')

# "ln_entities_per_cap ~ C(sex)*(age + children + C(smoker) + C(region)) + age:C(smoker)

model1 = linear_model.LinearRegression()

model1.fit(X, y)

mse = metrics.mean_squared_error(y, model1.predict(X))

print(model1.coef_)
```

```python
model1.intercept_[0]
```

```python
fig, ax = plt.subplots()

df.plot(x = "CPI_avg", y = "ln_entities_per_cap", kind = "scatter", ax=ax)

beta_1 = model1.coef_[0, 5]
beta_0 = model1.intercept_[0]

x = np.array([0, df["CPI_avg"].max()])
ax.plot(x, beta_0 + beta_1*x)
```

```python
df_small = df.dropna()

y, X = patsy.dmatrices("ln_entities_per_cap ~ CPI_avg + C(Region)", df_small, return_type='matrix')

# "ln_entities_per_cap ~ C(sex)*(age + children + C(smoker) + C(region)) + age:C(smoker)

model1 = linear_model.LinearRegression()

model1.fit(X, y)

mse = metrics.mean_squared_error(y, model1.predict(X))

print(mse)

print(model1.coef_)
```

```python
eqn = ("ln_entities_per_cap ~ C(Region) + CPI_avg + demox_eiu + gini_index"
"+ ln_income + invest_pct_gdp + tax_pct_gdp"
      )

y, X = patsy.dmatrices(eqn, df, return_type='matrix')

y = y.flatten()

model2CV = linear_model.LassoCV(cv=5).fit(X, y)

model2 = linear_model.Lasso(alpha = model2CV.alpha_)

model2.fit(X, y)

print(model2.coef_)

mse = metrics.mean_squared_error(y, model2.predict(X))

print(mse)

X;
```

```python
CPI["CPI_diff"] = CPI["CPI 2016 Score"] - CPI["CPI 2015 Score"]

df_new = pd.merge(df, CPI, left_on = "Code", right_on = "Country Code")

df_new = df_new.drop(columns = ["CPI 2016 Score", "CPI 2015 Score", "CPI 2014 Score",
                                "CPI 2013 Score", "CPI 2012 Score", "CPI_avg_y", "Region_y", "Country",
                                "Country Code", "CPI 2016 Rank"
                               ]
                    )
df_new = df_new.rename(columns = {"CPI_avg_x":"CPI_avg", "Country_x":"Country", "Region_x":"Region"})

df_new
```

```python
fig, ax = plt.subplots()

df_new.plot(x = "ln_entities_per_cap", y = "CPI_diff", kind = "scatter", ax=ax)
```

```python
eqn = ("CPI_diff ~ C(Region) + ln_entities_per_cap"
""
      )

y, X = patsy.dmatrices(eqn, df_new, return_type='matrix')

model3 = linear_model.LinearRegression()

model3.fit(X, y)

fig, ax = plt.subplots()

df_new.plot(y = "CPI_diff", x = "ln_entities_per_cap", kind = "scatter", ax=ax)

beta_1 = model3.coef_[0, 5]
beta_0 = model3.intercept_[0]

x = np.array([-5, df_new["ln_entities_per_cap"].min()])
ax.plot(x, beta_0 + beta_1*x)

mse = metrics.mean_squared_error(y, model3.predict(X))

print(mse)
```

```python
eqn = ("CPI_diff ~ C(Region) + ln_entities_per_cap:ln_income"
""
      )

y, X = patsy.dmatrices(eqn, df_new, return_type='matrix')

model4 = linear_model.LinearRegression()

model4.fit(X, y)


print(model3.coef_)
print(model4.coef_)

mse = metrics.mean_squared_error(y, model4.predict(X))

print(mse)
```
